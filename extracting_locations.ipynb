```python
# Import the necessary libraries

import geopandas as gpd
import geopy
import geocoder
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
geolocator = Nominatim(user_agent="Nancy Amandi", timeout= 10)
rgeocode = RateLimiter(geolocator.reverse, min_delay_seconds=0.1)
area["location"] = area["coordinates"].apply(rgeocode)
```

```python
area['location'][0]
```

```python
def get_suburb(row):
    location = row["location"]
    
    try:
        return location.raw["address"]["suburb"]
    except KeyError:
        return " "

def get_city(row):
    location = row["location"]
    try:
        return location.raw["address"]["city"]
    except KeyError:
        return " "

def get_state(row):
    location = row["location"]
    try:
        return location.raw["address"]["state"]
    except KeyError:
        return " "
    
```

```python
area["city"] = area.apply(get_suburb, axis=1)
area["suburb"] = area.apply(get_city, axis=1)
area["state"] = area.apply(get_state, axis=1)

area.head()
```
